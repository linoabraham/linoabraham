````python
%%writefile TareaAcademica.cpp
#include <iostream>
#include <string>
#include <cstdlib>
#include <ctime>
#include<fstream>
using namespace std;

class Persona
{
public:
	Persona(string n, string d, int t, int e, int nu):nombre(n),direccion(d),telefono(t),edad(e),numero(nu){}
	~Persona(){}
	void toString() {
		cout << nombre << " - " << direccion << " - " << telefono << " - " << edad << " - " << numero << endl;
	}
	int getNumero() {
		return this->numero;
	}
	int getTelefono() {
		return this->telefono;
	}
	int getEdad() {
		return this->edad;
	}
	string getNombre() {
		return this->nombre;
	}
	string getDireccion() {
		return this->direccion;
	}
private:
	string nombre;
	string direccion;
	int telefono;
	int edad;
	int numero;
};

class Nodo {
private:
	Nodo* siguiente;
	Persona* elemento;
public:
	Nodo(Persona* e) {
		elemento = e;
		siguiente = nullptr;
	}
	Persona* getElemento() {
		return elemento;
	}
	Nodo* getSiguiente() {
		return siguiente;
	}
	void setSiguiente(Nodo* s) {
		siguiente = s;
	}
};

class Lista {
private:
	Nodo* inicio;
	size_t tamanio;
public:
	Lista() {
		inicio = nullptr;
		tamanio = 0;
	}
	void insertar(Persona* e) {
		Nodo* nuevo = new Nodo(e);
		if (inicio == nullptr) 
			inicio = nuevo;
		else {
			Nodo* aux = inicio;
			while (aux->getSiguiente() != nullptr) 
				aux = aux->getSiguiente();
			aux->setSiguiente(nuevo);
			aux = nullptr;
		}
		tamanio++;
	}
	void mostrarLista() {
		Nodo* aux = inicio;
		cout << "Lista: " << endl;

		while (aux->getSiguiente() != nullptr) {
			aux->getElemento()->toString();
			aux = aux->getSiguiente();
		}
		aux = nullptr;
	}
	void ganadores(int pozo, int numero){
		Nodo* aux = inicio;
		int ganadores_6_digitos = 0;
		int ganadores_4_digitos = 0;
		int ganadores_Anterior_Siguiente = 0;
		bool existeGanador6digitos = false;
		bool existeGanador4digitos = false;

		while (aux != nullptr) {
			if (aux->getElemento()->getNumero() == numero) {
				++ganadores_6_digitos;
				existeGanador6digitos = true;
			}
			string strNumeroPersona = to_string(aux->getElemento()->getNumero());
			string strNumeroGanador = to_string(pozo);

			for (int i = 0; i < strNumeroPersona.size(); i++) {
				if (strNumeroPersona[i] == strNumeroGanador[i] && strNumeroPersona[i + 1] == strNumeroGanador[i + 1] && strNumeroPersona[i + 2] == strNumeroGanador[i + 2] && strNumeroPersona[i + 3] == strNumeroGanador[i + 3]) {
					++ganadores_4_digitos;
					existeGanador4digitos = true;
				}
			}

			if (existeGanador4digitos || existeGanador6digitos) {
				++ganadores_Anterior_Siguiente;
			}

			existeGanador6digitos = false;
			existeGanador4digitos = false;
			aux = aux->getSiguiente();
		}

		cout << endl <<"---------------Reporte de Ganadores--------------" << endl;
		cout << "Ganadores de 6 digitos: " << ganadores_6_digitos << endl;
		cout << "Ganadores de 4 digitos: " << ganadores_4_digitos << endl;
		cout << "Ganadores que compraron su numero justo despues de un ganador: " << ganadores_Anterior_Siguiente << endl;
		cout << "Cada ganador de 6 digitos recibira: "; if (existeGanador6digitos > 0) { cout << pozo / existeGanador6digitos; } cout << " soles" << endl;
		cout << "Cada ganador de 4 digitos recibira: " << 20000 << " soles" << endl;
		cout << "Cada ganador que compro su numero justo despues de un ganador recibira: "; if (existeGanador4digitos > 0) { cout << (pozo / existeGanador6digitos) * 0.1; } cout << " soles";
	}
	void crearArchivo() {
		ofstream fs("listaParticipantes.txt");
		Nodo* aux = inicio;
		while (aux != nullptr) {
			fs << "Nombre: " << aux->getElemento()->getNombre() << endl;
			fs << "Direccion: " << aux->getElemento()->getDireccion() << endl;
			fs << "Telefono: " << aux->getElemento()->getTelefono() << endl;
			fs << "Edad: " << aux->getElemento()->getEdad() << endl;
			fs << "Numero: " << aux->getElemento()->getNumero() << endl;
			aux = aux->getSiguiente();
		}
		fs.close();
	}
};

int main() {
	srand((unsigned)time(0));
	Lista* l = new Lista();
	for (int i = 0; i < 5; i++)
		l->insertar(new Persona("aaaa", "bbbb", ((rand() % 9999999) + 1000000), ((rand() % 90) + 18), ((rand() % 999999) + 100000)));
	l->mostrarLista();
	l->ganadores(50000, 123456);
	l->crearArchivo();
	cin.get();
	cin.ignore();
	return 0;
}


````
